```python
transcriptome_cpm_all_expressed_genes = pd.read_csv("00 DATA AND FINAL ANALYSIS RKO PKO RPKO.csv")
transcriptome_cpm_all_expressed_genes
```

## CPM

```python
# transcriptome_cpm_all_expressed_genes.iloc[:,1:3]
columns = transcriptome_cpm_all_expressed_genes.columns
num_col = len(transcriptome_cpm_all_expressed_genes.columns)
cpm = transcriptome_cpm_all_expressed_genes.copy()
for i in range(0,round(num_col/2)):
    cpm.loc[:,columns[i*2+1]] =transcriptome_cpm_all_expressed_genes.loc[:,[columns[i*2+1],columns[i*2+2]]].sum(axis=1)/transcriptome_cpm_all_expressed_genes.loc[:,[columns[i*2+1],columns[i*2+2]]].sum(axis=0).sum()*1000000
    print(cpm.loc[:,columns[i*2+1]].sum())
    cpm = cpm.drop([columns[i*2+2]],axis=1)
cpm
```

## Normalize row

```python
# for testing 
# transcriptome_cpm_all_expressed_genes = transcriptome_cpm_all_expressed_genes.iloc[0:100,:]

num_col = len(cpm.columns)

# transcriptome_cpm_all_expressed_genes.iloc[:,1:num_col].div(transcriptome_cpm_all_expressed_genes.iloc[:,1:num_col].sum(axis=1), axis=0)
for i in range(0,len(cpm)):
    x = cpm.iloc[i,1:num_col]
#     transcriptome_cpm_all_expressed_genes.iloc[i,1:num_col] = x/ np.linalg.norm(x)
    cpm.iloc[i,1:num_col] = x/ np.amax(x)
cpm
```
