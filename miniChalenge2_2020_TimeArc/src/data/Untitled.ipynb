```python
Labels = pd.DataFrame({'person_as_time':[],'img':[],'term':[],'category':[],'score':[]})
start_time = pd.to_datetime('1/1/2019')
for file_name in glob.iglob('./Manual Correction Results/*.csv', recursive=True):
    personIndex = file_name.split('\\')[-1].replace('Person','').replace('_manual.csv','')
    time = start_time + pd.to_timedelta(personIndex+' days')
    df = pd.read_csv(file_name,encoding = "ISO-8859-1")
    term = []
    category = []
    score = []
    for index in range(0,df.shape[0]):
        row = df.loc[index]
        img = 'Person'+personIndex+'_'+str(row['Image'])
        for col in df.columns:
            if (col != 'Image'):
                label = row[col]
                try:
                    math.isnan(label)
                except:
                    term.append(label)
                    category.append('label')
                    score.append(label+':'+str(1))
    Labels = Labels.append({'person_as_time':time,'img':img,'term':'|'.join(term),'category':'|'.join(category),'score':';'.join(score)},ignore_index=True)
```

```python
Labels.to_csv('data_fixed.csv',index=False)
```
