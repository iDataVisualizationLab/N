```python
Labels = pd.DataFrame({'person_as_time':[],'img':[],'term':[],'category':[],'score':[]})
start_time = pd.to_datetime('1/1/2019')
for file_name in glob.iglob('./**/*.csv', recursive=True):
    personIndex = file_name.split('\\')[-2].replace('Person','')
    time = start_time + pd.to_timedelta(personIndex+' days')
    df = pd.read_csv(file_name,encoding = "ISO-8859-1")
    img = file_name.split('\\')[-1].split('.csv')[0]
    term = []
    category = []
    score = []
    for index in range(0,df.shape[0]):
        row = df.loc[index]
        label = row['Label']
        term.append(label)
        category.append('label')
        score.append(label+':'+str(row['Score']))
    Labels = Labels.append({'person_as_time':time,'img':img,'term':'|'.join(term),'category':'|'.join(category),'score':';'.join(score)},ignore_index=True)
```

```python
Labels.to_csv('data.csv',index=False)
```
