```python
Labels = pd.DataFrame({'Person':[],'Image':[],'Label':[],'Score':[]})
for file_name in glob.iglob('./Manual Correction Results/*.csv', recursive=True):
    personIndex = file_name.split('\\')[-1].replace('Person','').replace('_manual.csv','')
    df = pd.read_csv(file_name,encoding = "ISO-8859-1")
    for index in range(0,df.shape[0]):
        row = df.loc[index]
        img = 'Person'+personIndex+'_'+str(row['Image'])
        for col in df.columns:
            if (col != 'Image'):
                label = row[col]
                try:
                    math.isnan(label)
                except:
                    Labels = Labels.append({'Person':personIndex,'Image':img,'Label':label,'Score':1},ignore_index=True)
```

```python
Labels.to_csv('..\..\..\miniChalenge2_2020_ViolinMatrix\src\data\data_manualCorrect.csv',index=False)
```
